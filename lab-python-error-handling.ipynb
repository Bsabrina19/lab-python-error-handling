````python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

# Or, in another way:

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
````

````python
def calculate_total_price(customer_orders):
    """Prompts the user to enter valid prices for ordered products and calculates total price."""
    total_price = 0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    print("Error: Price cannot be negative. Please enter a valid amount.")
                else:
                    total_price += price * customer_orders[product]  # Multiply by ordered quantity
                    valid_price = True
            except ValueError:
                print("Invalid input. Please enter a numeric price.")
    
    return total_price
````

````python
def get_customer_orders(products, inventory):
    """Prompts the user for valid product orders, ensuring correct input and stock availability."""
    
    # Validate the number of orders
    while True:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders < 0:
                print("Error: The number of orders cannot be negative. Please enter a valid number.")
            else:
                break  # Exit loop when valid input is entered
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    customer_orders = {}

    for _ in range(num_orders):
        valid_product = False
        while not valid_product:
            product = input(f"Enter the name of a product to order (choose from: {products}): ").strip().lower()

            if product not in inventory:
                print("Error: Product does not exist. Please enter a valid product name.")
            elif inventory[product] <= 0:
                print(f"Error: {product} is out of stock. Please choose another product.")
            else:
                valid_product = True

        # Validate the quantity of the order
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product} to order: "))
                if quantity < 0:
                    print("Error: Quantity cannot be negative. Please enter a valid number.")
                elif quantity > inventory[product]:
                    print(f"Error: Not enough stock available! Only {inventory[product]} units left.")
                else:
                    customer_orders[product] = quantity
                    valid_quantity = True
            except ValueError:
                print("Invalid input. Please enter a whole number.")

    return customer_orders
````

````python
products=["t-shirt","mug","hat","book","keychain"]

inventory=initialize_inventory(products)
print(f'available stock is:\n {inventory}')

customer_orders=get_customer_orders(products, inventory)
print(f'Customer order list is:\n{customer_orders}')

total_price=calculate_total_price(customer_orders)
print(f'the total price of the customer orders is :\n {total_price}£')
````
